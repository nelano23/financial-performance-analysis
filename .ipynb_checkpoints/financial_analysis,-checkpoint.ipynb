# Financial Sales Performance Analysis

## Overview
Analysis of 10,800+ Superstore transactions to identify regional and category-level sales and profit trends using Python and pandas.

## Tools Used
- Python
- pandas
- matplotlib
- Jupyter Notebook

## Key Findings
1. The West region leads in both sales and profit, while Central has surprisingly low profit despite being the second highest in sales.
2. Furniture generates high sales but almost no profit compared to Technology and Office Supplies.
3. Technology is the most profitable category, making nearly 8x more profit than Furniture despite similar sales volume.

## Dataset
Superstore sales data — 9,994 transactions across 21 columns including sales, profit, region, and product category.

```python
import pandas as pd
df = pd.read_csv('/Users/nelloeanonyuo/Desktop/superstore.csv')
df.head()
```

```python
df.shape #get row and column number of df 
```

```python
df.columns #get column names 
```

```python
print(df.info()) #prints out more info about the df 
```

```python
df.isnull().sum() #sum of all null values
```

```python
df = df.dropna(subset=['Sales', 'Profit']) #Dropping rows where sales and profit are null
df.shape
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Region')['Profit'].sum()
```

```python
df.groupby('Category')[['Sales', 'Profit']].sum()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Region')
plt.ylabel('Sales')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title('Total Profit by Category')
plt.ylabel('Profit')
plt.xlabel('Category')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title('Total Profit by Region')
plt.ylabel('Profit')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

Major Findings from Analysis:
1. The West region leads in both sales and profit, while Central has surprisingly low profit despite being the second highest in sales.
2. Furniture generates high sales but almost no profit compared to Technology and Office Supplies.
3. Technology is the most profitable category, making nearly 8x more profit than Furniture despite similar sales volume.
